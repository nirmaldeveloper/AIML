#### 2: Write a user-defined function for origin  


```python
#use apply function
def Replace_Origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    elif num == 3:
        return 'Asia'

df_auto_data['origin'] = df_auto_data['origin'].apply(Replace_Origin)

df_auto_data

        
    
```
